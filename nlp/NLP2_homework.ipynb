```python
#NLP2_3
from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()

data = list(open("./labeled.csv", encoding="utf-8"))

data_tok = []
for i in range(len(data)): 
  data_tok.append(tokenizer.tokenize(data[i].lower()))
```

```python
data_tok[5]
```

```python
sl = {}
for i in data_tok:
  for j in i:
    if j not in sl:
      sl[j]=1
    else:
      sl[j]+=1

def vectorize(tokens):
    vector=[]
    for w in sl:
        vector.append(tokens.count(w))
    return vector

print(vectorize(data_tok[0][:15]))
```
