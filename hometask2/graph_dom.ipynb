```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_dom = pd.read_csv('domrf.csv')
df_dom.head()
```

```python
df_dom = df_dom.loc[:, df_dom.isnull().mean() < .3]
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df_dom.head(1).T
```

```python
df_dom.head(1)
```

```python
df_dom.objTransferPlanDt = pd.to_datetime(df_dom.objTransferPlanDt).dt.date
```

```python
df_dom = df_dom[['id','region','floorMax','objElemLivingCnt','objReady100PercDt','objTransferPlanDt','wallMaterialShortDesc','objLkFinishTypeDesc', 'objLkFreePlanDesc',\
                'objElemParkingCnt', 'objLkClassDesc', 'objFlatCnt', 'objFlatSq', 'objInfrstrBicycleLaneFlg', 'objInfrstrPlaygrndCnt',\
                'objInfrstrTrashAreaCnt', 'objInfrstrRampFlg', 'objInfrstrCurbLoweringFlg', 'objElevatorWheelchairCnt', 'objElevatorPassengerCnt', 'objElevatorCargoCnt',\
                'objPriceAvg', 'nonlivFirstFloor', 'objEnergyEffShortDesc', 'infrastructureIndexValue']]
```

```python
df_dom.head(5)
```

```python
df_dom['objReady100PercDt_month'] = pd.DatetimeIndex(df_dom['objReady100PercDt']).month
df_dom['objTransferPlanDt_month'] = pd.DatetimeIndex(df_dom['objTransferPlanDt']).month
df_dom['objReady100PercDt_year'] = pd.DatetimeIndex(df_dom['objReady100PercDt']).year
```

```python
Quarter_Mapping = {1:1, 2:1, 3:1, 4:2, 5:2, 6:2, 7:3, 8:3, 9:3, 10:4, 11:4, 12:4}
df_dom['objReady100PercDt_Quarter'] = df_dom['objReady100PercDt_month'].map(Quarter_Mapping)
```

```python
df_dom['objTransferPlanDt'].isnull().sum()
```

```python
df_dom = df_dom.drop_duplicates(keep='last')
```

```python
df_dom.dropna(subset=['objTransferPlanDt'],inplace=True)
```

```python
df_dom.head()
```

```python
df_dom.dropna(subset=['objPriceAvg'],inplace=True)
```

```python
df_dom.isnull().sum()
```

```python
df_dom.fillna(value=0, inplace=True)
```

```python
#С каждым годом кол-во кв.м. увеличивается, спад в 2027г. связан, возможно с тем,
#что еще не все строящиеся дома зарегистрированы со сроком ввода в эксплуатацию >= 2027г'
sns.relplot(
    x='objReady100PercDt_year', 
    y='objFlatSq', 
    data=df_dom.query("region == 77"), 
    kind='line',ci=None,height=10
)
plt.title('Ввод в эксплуатацию', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
#Меньше всего кв.м. элитного класса недвижимости, заметный рост кв.м. бизнес класса к 2026г. 
sns.relplot(x='objReady100PercDt_year', y='objFlatSq', hue='objLkClassDesc',
            data=df_dom.query("region == 77"), kind='line',ci=None,height=10)
plt.title('Ввод в эксплуатацию (по классу недвижимости)', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
#Больше всего строится недвижимостит со свободной планировкой
sns.relplot(x='objReady100PercDt_year', y='objFlatSq', hue='nonlivFirstFloor', style = 'objLkFreePlanDesc',
            dashes=True, markers=True,
            data=df_dom.query("region == 77"), kind='line',err_style="bars", ci=None,height=10)
plt.title('Ввод в эксплуатацию (жилой 1-й этаж и свободная планировка)', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
sns.relplot(
    x='objReady100PercDt_year', 
    y='objFlatSq', 
    data=df_dom, 
    kind='line', 
    hue='objLkClassDesc', 
    row='nonlivFirstFloor', 
    col='objLkFreePlanDesc',ci=None,height=8);
```

```python
#строяится недвижимость от 2 до ~35 этажей
sns.relplot(x='objReady100PercDt_year', y='floorMax', data=df_dom, kind='scatter',height=7)
plt.title('Ввод в эксплуатацию по этажности недвижимости', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Этажность', size=15, color='g');
```

```python
# Недвижимость в основном без свободной планировки
sns.relplot(x='objReady100PercDt_year', y='floorMax', hue='objLkFreePlanDesc', data=df_dom,height=7)
plt.title('Ввод в эксплуатацию по этажности недвижимости по типу планировки', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Этажность', size=15, color='g');
```

```python
sns.relplot(x='objReady100PercDt_year', y='floorMax', hue='nonlivFirstFloor', style='objLkFreePlanDesc', markers=['^', 'D'],
            data=df_dom,height=10)
plt.title('Ввод в эксплуатацию по этажности недвижимости свободная планировка (жилой 1-й этаж и свободная планировка)', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Этажность', size=15, color='g');
```

```python
#Больше всего недвижимости с классом энергоэффективности B,A и С
sns.relplot(
    x='objReady100PercDt_year',
    y='objPriceAvg',
    hue='objLkFreePlanDesc',
    col='objEnergyEffShortDesc',
    col_wrap=3,
    data=df_dom,
    height=5
);
```

```python
sns.catplot(
    x='objLkClassDesc',
    y='objFlatSq', 
    data=df_dom, 
    palette='Set1', 
    jitter=True,
    height=4,
    aspect=3); 

plt.title('stripplot с параметром jitter, заданным по умолчанию', size=20, color='g');
```

```python
sns.set(style='darkgrid')
sns.catplot(
    x='objLkClassDesc', 
    y='objFlatSq', 
    data=df_dom, 
    kind='box', 
    height=8, 
    aspect=1
); 
```

```python
sns.catplot(
    x='objLkClassDesc', 
    y='objFlatSq', 
    data=df_dom, 
    kind='box', 
    hue='objLkFreePlanDesc',
    height=8, 
    aspect=1); 
```

```python
sns.catplot(
    x='objLkFinishTypeDesc', 
    y='objFlatSq', 
    data=df_dom, 
    kind='boxen', 
    height=8, 
    aspect=1); 
```

```python
sns.catplot(x='objLkClassDesc', 
            data=df_dom, 
            kind='count', 
            height=7, 
            aspect=2
           );  # long form data frame
plt.title('Кол-во недвижимости по классу', size=20, color='g', weight='bold');
```

```python
sns.catplot(
    x='objLkFinishTypeDesc', 
    data=df_dom, 
    kind='count', 
    hue='objLkFreePlanDesc',
        hue_order=['Нет','Есть'], 
    palette={"Нет": "r", "Есть": "g"},
    height=4, 
    aspect=2); 
plt.title('6. Count Plot, hue=Promotion, custom palette', size=15, color='g', weight='bold');
```

```python
sns.catplot(
    x='objLkClassDesc', 
    y='objFlatSq', 
    data=df_dom, 
    kind='violin', 
    hue='objLkFreePlanDesc', 
    hue_order=['Нет','Есть'], 
    split=True, 
    height=7, 
    aspect=2);
```

```python
sns.catplot(
    x='objLkFinishTypeDesc', 
    y='objFlatSq', 
    data=df_dom, # long form data frame
    kind='point', 
    markers='D', 
    ci='sd',
    height=4, 
    aspect=2)  
plt.title('2.  Точечная диаграмма с алмазным маркером и SD в доверительном интервале', size=15, color='g', weight='bold');
```
