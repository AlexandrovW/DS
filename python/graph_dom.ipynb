```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_dom = pd.read_csv('domrf.csv')
df_dom.head()
```

```python
df_dom = df_dom.loc[:, df_dom.isnull().mean() < .3]
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df_dom.head(1).T
```

```python
df_dom.objTransferPlanDt = pd.to_datetime(df_dom.objTransferPlanDt).dt.date
```

```python
df_dom = df_dom[['id','region','floorMax','objElemLivingCnt','objReady100PercDt','objTransferPlanDt','wallMaterialShortDesc','objLkFinishTypeDesc', 'objLkFreePlanDesc',\
                'objElemParkingCnt', 'objLkClassDesc', 'objFlatCnt', 'objFlatSq', 'objInfrstrBicycleLaneFlg', 'objInfrstrPlaygrndCnt',\
                'objInfrstrTrashAreaCnt', 'objInfrstrRampFlg', 'objInfrstrCurbLoweringFlg', 'objElevatorWheelchairCnt', 'objElevatorPassengerCnt', 'objElevatorCargoCnt',\
                'objPriceAvg', 'nonlivFirstFloor', 'objEnergyEffShortDesc', 'infrastructureIndexValue']]
```

```python
df_dom['objReady100PercDt_month'] = pd.DatetimeIndex(df_dom['objReady100PercDt']).month
df_dom['objTransferPlanDt_month'] = pd.DatetimeIndex(df_dom['objTransferPlanDt']).month
df_dom['objReady100PercDt_year'] = pd.DatetimeIndex(df_dom['objReady100PercDt']).year
```

```python
Quarter_Mapping = {1:1, 2:1, 3:1, 4:2, 5:2, 6:2, 7:3, 8:3, 9:3, 10:4, 11:4, 12:4}
df_dom['objReady100PercDt_Quarter'] = df_dom['objReady100PercDt_month'].map(Quarter_Mapping)
```

```python
df_dom['objTransferPlanDt'].isnull().sum()
```

```python
df_dom = df_dom.drop_duplicates(keep='last')
```

```python
df_dom.dropna(subset=['objTransferPlanDt'],inplace=True)
```

```python
df_dom.dropna(subset=['objPriceAvg'],inplace=True)
```

```python
df_dom.isnull().sum()
```

```python
df_dom.fillna(value=0, inplace=True)
```

```python
df_dom = df_dom.rename(columns={'objReady100PercDt_year': 'Год сдачи объекта', 'objFlatSq':'Площадь кв.м.', 'region':'Регион',
                               'objLkClassDesc':'Класс жилья', 'nonlivFirstFloor': 'Не жилой первый этаж', 'objLkFreePlanDesc': 'Свободная планировка',
                               'floorMax': 'Кол-во этажей','objEnergyEffShortDesc': 'Класс энергоэффективности здания','objPriceAvg': 'Средняя цена за кв.м.',
                               'objLkFinishTypeDesc': 'Тип отделки'})
```

```python
df_dom['Не жилой первый этаж'] = df_dom['Не жилой первый этаж'].replace([1,0],['Да','Нет'])
```

```python
#С каждым годом кол-во кв.м. увеличивается, спад в 2027г. связан, возможно с тем,
#что еще не все строящиеся дома зарегистрированы со сроком ввода в эксплуатацию >= 2027г'
sns.relplot(
    x='Год сдачи объекта', 
    y='Площадь кв.м.', 
    data=df_dom.query("Регион == 77"), 
    kind='line',ci=None,height=10
)
plt.title('Ввод в эксплуатацию', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
#Меньше всего кв.м. элитного класса недвижимости, заметный рост кв.м. бизнес класса к 2026г. 
sns.relplot(x='Год сдачи объекта', y='Площадь кв.м.', hue='Класс жилья',
            data=df_dom.query("Регион == 77"), kind='line',ci=None,height=10)
plt.title('Ввод в эксплуатацию (по классу недвижимости)', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
#Больше всего строится недвижимостит со свободной планировкойи не жилым 1 этажом
sns.relplot(x='Год сдачи объекта', y='Площадь кв.м.', hue='Не жилой первый этаж', style = 'Свободная планировка',
            dashes=True, markers=True,
            data=df_dom.query("Регион == 77"), kind='line',err_style="bars", ci=None,height=10)
plt.title('Ввод в эксплуатацию (жилой 1-й этаж и свободная планировка)', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
sns.relplot(
    x='Год сдачи объекта', 
    y='Площадь кв.м.', 
    data=df_dom, 
    kind='line', 
    hue='Класс жилья', 
    row='Не жилой первый этаж', 
    col='Свободная планировка',ci=None,height=8);
```

```python
#строяится недвижимость от 2 до ~35 этажей
sns.relplot(x='Год сдачи объекта', y='Кол-во этажей', data=df_dom, kind='scatter',height=7)
plt.title('Ввод в эксплуатацию по этажности недвижимости', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кол-во этажей', size=15, color='g');
```

```python
# Недвижимость в основном без свободной планировки
sns.relplot(x='Год сдачи объекта', y='Кол-во этажей', hue='Свободная планировка', data=df_dom,height=7)
plt.title('Ввод в эксплуатацию по этажности недвижимости по типу планировки', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кол-во этажей', size=15, color='g');
```

```python
#Чем выше дом, тем больше не жилых 1 этажей
sns.relplot(x='Год сдачи объекта', y='Кол-во этажей', hue='Не жилой первый этаж', style='Свободная планировка', markers=['^', 'D'],
            data=df_dom,height=8, aspect =2)
plt.title('Ввод в эксплуатацию по этажности недвижимости свободная планировка (жилой 1-й этаж и свободная планировка)', size=20, color='g')
plt.xlabel('Год', size=15, color='g')
plt.ylabel('Кол-во этажей', size=15, color='g');
```

```python
#Больше всего недвижимости с классом энергоэффективности B,A и С
sns.relplot(
    x='Год сдачи объекта',
    y='Площадь кв.м.',
    hue='Свободная планировка',
    col='Класс энергоэффективности здания',
    col_wrap=3,
    data=df_dom,
    height=5
);
```

```python
#Большой разброс в цене за кв.м. в элитном классе жилья
sns.catplot(
    x='Класс жилья',
    y='Средняя цена за кв.м.', 
    data=df_dom, 
    palette='Set1', 
    jitter=True,
    height=4,
    aspect=3); 

plt.title('Стоимость жилья по классу', size=20, color='g')
plt.xlabel('Класс жилья', size=15, color='g')
plt.ylabel('Средняя цена за кв.м.', size=15, color='g');
```

```python
#Много выбросов по классу жилья Бизнес
sns.set(style='darkgrid')
sns.catplot(
    x='Класс жилья', 
    y='Площадь кв.м.', 
    data=df_dom, 
    kind='box', 
    height=8, 
    aspect=1
)
plt.title('Объем жилья по классу', size=20, color='g')
plt.xlabel('Класс жилья', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
#В элитном жилье заметно больше строят без свободной планировки
sns.catplot(
    x='Класс жилья', 
    y='Площадь кв.м.', 
    data=df_dom, 
    kind='box', 
    hue='Свободная планировка',
    height=8, 
    aspect=1)
plt.title('Объем жилья по классу и свободной планировке', size=20, color='g')
plt.xlabel('Класс жилья', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g');
```

```python
sns.catplot(
    x='Тип отделки', 
    y='Площадь кв.м.', 
    data=df_dom, 
    kind='boxen', 
    height=8, 
    aspect=1)
plt.title('Объем жилья по типу отделки', size=20, color='g')
plt.xlabel('Класс жилья', size=15, color='g')
plt.ylabel('Кв.М.', size=15, color='g'); 
```

```python
#Больше всего строится недвижимости класса комфорт
sns.catplot(x='Класс жилья', 
            data=df_dom, 
            kind='count', 
            height=7, 
            aspect=2
           )
plt.title('Кол-во недвижимости по классу', size=20, color='g', weight='bold')
plt.xlabel('Класс жилья', size=15, color='g')
plt.ylabel('Кол-во.', size=15, color='g');
```

```python
#Свободная планировка предоставляется в основном без отделки
sns.catplot(
    x='Тип отделки', 
    data=df_dom, 
    kind='count', 
    hue='Свободная планировка',
        hue_order=['Нет','Есть'], 
    palette={"Нет": "r", "Есть": "g"},
    height=8, 
    aspect=2) 
plt.title('Недвижимость по типу отделки и свободной планировки', size=20, color='g', weight='bold')
plt.xlabel('Тип отделки', size=15, color='g')
plt.ylabel('Кол-во.', size=15, color='g');
```

```python
sns.catplot(
    x='Класс жилья', 
    y='Площадь кв.м.', 
    data=df_dom, 
    kind='violin', 
    hue='Свободная планировка', 
    hue_order=['Нет','Есть'], 
    split=True, 
    height=7, 
    aspect=2)
plt.title('Недвижимость по классу жилья и свободной планировки', size=20, color='g', weight='bold')
plt.xlabel('Класс жилья', size=15, color='g')
plt.ylabel('Кв.м.', size=15, color='g');
```
